# Executive summary

We set out to investigate whether the increasing adoption of electric vehicles (EVs) in Switzerland has led to measurable improvements in air quality, focusing specifically on nitrogen oxide (NOx) and carbon monoxide (CO) levels. Using monthly data from 2006 onward across seven representative cantons, we conducted a thorough correlational analysis.<br><br>Our results show that while EV uptake, especially after 2018, coincides with continued declines in NOx and CO levels, these improvements had already begun earlier, and there is no clear evidence that the decline accelerated post-2018.

Linear, polynomial, Lasso, and Ridge regressions, along with a classification analysis, all confirmed that EV adoption explains only a small fraction of the variation in air quality. This suggests that other factors are more likely responsible for the observed improvements
